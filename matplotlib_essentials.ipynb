```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

# creating plot

```python
x = np.linspace(0, 10)
y = np.cos(x)
plt.plot(x, y)
```

```python
plt.show(plt.plot(x - 10, y))
```

```python
z = np.sin(x)
plt.plot(x,y,x,z)
plt.show()
```

```python
plt.xlabel('Time')
plt.ylabel('Magnitude')
plt.title('Sin & Cos waves')
plt.plot(x,y,x,z)
plt.legend(['y','z'], loc=0)
plt.show()
```

## subplotting 

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'g-o')
plt.title('Cosine')

plt.subplot(1,2,2)
plt.plot(x,z, 'r-x')
plt.title('Sine')

plt.tight_layout()
plt.show()
```

```python
x = np.arange(0,11)
y1 = x ** 2
y2 = 2 * x ** 2 + 3 * x
plt.plot(x, y1, 'b-o', x, y2, 'r-o')
plt.xlabel('argument')
plt.ylabel('magnitude')
plt.title('Plots of X')
```
